```python
file_list=["HFFc6-FA-DpnII-HFF_strong_extended.bed",
"HFFc6-DSG-DpnII-HFF_strong_extended.bed",
"HFFc6-FA-DSG-MNase-HFF_strong_extended.bed",
"ESC4DN-FA-DpnII-ESC_strong_extended.bed",
"ESC4DN-DSG-DpnII-ESC_strong_extended.bed",
"H1ESC4DN-FA-DSG-MNase-ESC_strong_extended.bed"]
x_name=["FA-DpnII-Deep","DSG-DpnII-Deep","DSG-MNase-Deep"]*2
y_name=["FA-DpnII","DSG-DpnII","DSG-MNase"]*2
```

```python
lw = 2
from sklearn import linear_model, datasets

path="/nl/umw_job_dekker/users/ba69w/HiC_Analysis/U54_matrix/insulation_sameer_v2/window_200000/10000_20px_v3/"

for i,file in enumerate(file_list):
    file1 = pd.read_table(path+file, sep="\t", index_col=None,header=None)
    #print(file1.head(5))
    name=file.split(".")[0]
    #print(name)
    #shifted_scores=file1[file1.iloc[:,5]!=file1.iloc[:,11]]
    #shifted_scores.head(5)
    col1 = file1.iloc[:,3]
    col2 = file1.iloc[:,7]
    X = col1.values.reshape(-1,1)
    y = col2.values.reshape(-1,1)
    lr = linear_model.LinearRegression()
    lr.fit(X, y)
    res=np.round(lr.score(X, y),2)
    #print(str(res))
    line_X = np.arange(X.min(), X.max())[:, np.newaxis]
    line_y = lr.predict(line_X)
    fig, ax = plt.subplots(ncols=1,sharex=True, sharey=True,figsize=(5,5))
    ax.scatter(col1,col2,color='r', s=7, marker="o")
    #ax.scatter(shifted_scores.iloc[:,3],shifted_scores.iloc[:,9],color='r', s=7, marker="o")
    ax.set_xlabel(x_name[i],fontsize=30)
    ax.set_ylabel(y_name[i],fontsize=30)
    ax.plot(line_X, line_y, color='k', linewidth=lw, label=res)
    ax.set_xlim([0,4])
    ax.set_ylim([0,4])
    plt.legend(fontsize=20)
    plt.savefig(name+".pdf")
```

```python
def pile_up(df,index_list):
    start = df.ix[index_list].index-1 
    end  = df.ix[index_list].index+2
    #print(end)
    return (np.nanmean(np.array(list(map(lambda x: df["log2_insulation_score_200000"].values[x[0]:x[1]],  zip(start,end)))), axis=0))


```
