```python
hg38 = bioframe.fetch_chromsizes('hg38')
chromsizes = bioframe.fetch_chromsizes('hg38')
chromosomes = list(chromsizes.index)

print(filename1,filename2,filename3,filename4)

conditions = ['FA_MNase','FA_DdeI','FA_DpnII','FA_HindIII']


cool = [filename1,filename2,filename3,filename4]


cooler_paths = {    
    'FA_MNase' : path+filename1+'::/resolutions/100000',
    'FA_DdeI' : path+filename2+'::/resolutions/100000',
    'FA_DpnII' : path+filename3+'::/resolutions/100000',
    'FA_HindIII' : path+filename4+'::/resolutions/100000',
    
}

tracks = {    
    'FA_MNase': path1+track1,
    'FA_DdeI' :path1+track2 ,
    'FA_DpnII': path1+track3,
    'FA_HindIII': path1+track4,
    }
    

long_names = {
    'FA_MNase': 'FA_MNase',
    'FA_DdeI' :'FA_DdeI' ,
    'FA_DpnII': 'FA_DpnII',
    'FA_HindIII': 'FA_HindIII',
}

clrs = {
    cond: cooler.Cooler(cooler_paths[cond]) for cond in conditions
}


```

```python
fig = plt.figure(figsize=(20,10))
gs = gridspec.GridSpec(nrows=1,ncols=4)
k=0
for k,cond in enumerate(conditions):
    #print(track.shape)
    ax = plt.subplot(gs[k])
    x,y=clrs[cond].extent("chr12")
    #x=x+300
    #y=x+300
    #print(x,y)
    #print(y-x)
    x1=x+400
    y1=x+800
    #print(x1,y1)
    mat = clrs[cond].matrix(balance=True, sparse=True)[x1:y1,x1:y1]
    arr = mat.toarray()
    im =ax.matshow(np.log10(arr), cmap='YlOrRd',vmin=-4,vmax=-2.5)
    ax.set_title(long_names[cond], fontdict={'fontsize': 20, 'fontweight': 'medium'})
    #plt.colorbar(im)
    k=k+1
#plt.colorbar(im)
#plt.savefig('HFF_chr12_100kb_400_800.pdf')
```
